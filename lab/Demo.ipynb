```python
from gantt import Task, TaskScheduler, tasks_from_map
import datetime
import json
```

```python
resources = {'x': 2, 'y': 3}
```

```python
tasks = tasks_from_map(json.loads(tasks_basic))
schedule = TaskScheduler(datetime.datetime.now(), resources, tasks)
```

```python
schedule.schedule()
```

```python
for t in schedule.tasks:
    print(t)
```
